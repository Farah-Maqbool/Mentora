```python
import os
from dotenv import load_dotenv
from llama_index.llms.groq import Groq
from retriever import guest_info_tool
from tools import search_tool, weather_tool, news_tool
from llama_index.core.agent.workflow import AgentWorkflow

load_dotenv()


llm = Groq(model='openai/gpt-oss-120b', api_key=os.getenv('GROQ_API_KEY'), is_streaming=False)

alferd = AgentWorkflow.from_tools_or_functions(
    [guest_info_tool, search_tool, weather_tool, news_tool],
    llm=llm
)

query = "Tell me about Lady Ada Lovelace. What's her background?"
reponse = await alferd.run(query)
print(reponse.response.blocks[0].text)
```
